## code from scrapping from website

```python
Title=[]
Rating=[]
Price=[]
Battery = []
Brand = []
Storage=[]
Specification=[]
Warranty =[]
Displaylength=[]
for i in range(1,41):
    url = f'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=d257821d-8b45-442e-ba31-5487fa2d3564&page={i}'
    page = requests.get(url)
    
    soup = BeautifulSoup(page.text)
#Title
    for i in soup.find_all('div',class_='_2kHMtA'):
        title=i.find('div',class_='_4rR01T').text
        if title:
            Title.append(title)
        else:
            Title.append(np.nan)
            
#rating
    for i in soup.find_all('div','_2kHMtA'):
        rating = i.find('div', class_='_3LWZlK')
        if rating:
            Rating.append(rating.text)
        else:
            Rating.append(np.nan)
        
#price            
    for i in soup.find_all('div',class_='_2kHMtA'):
        price = i.find('div',class_='_30jeq3 _1_WHN1')
        if price:
            Price.append(price.text)
        else:
            Price.append(np.nan)
            
#specifications    
    for i in soup.find_all('div',class_='_2kHMtA'):
        spec = i.find('ul',class_='_1xgFaf').text
        if spec:
            Specification.append(spec)
        else:
            Specification.append(np.nan)   
```

## Extracting the details using regex

```python
df = pd.read_csv(r"C:\Users\CHARISHMA\projects\filedataframe.csv")
```

```python
#Brand
regex = '^\w+'
df['Title'].apply(lambda x: re.findall(regex,x))
df['Brand']=df['Title'].apply(lambda x: re.findall(regex,x))
```

```python
df.head()
```

```python
df['Specifications'][0]
```

```python
#length in cms
regex = 'd\d.\d\d.cm|\d+ cm'
df['Specifications'].apply(lambda x :re.findall(regex,x))
df['Length_in_cm']=df['Specifications'].apply(lambda x:re.findall(regex,x))
```

```python
df.head()
```

```python
#length in inches
regex = '\d.\d.inch|d+.\d+ inch|\d.\d\d inch|\d+ inch|\d+.\d+ Inch'
df['Specifications'].apply(lambda x :re.findall(regex,x))
df['Length_in_inches']=df['Specifications'].apply(lambda x:re.findall(regex,x))
```

```python
#Battery
regex='\d\d+ mAh '
df['Specifications'].apply(lambda x : re.findall(regex,x))
df['Battery_in_mAh']=df['Specifications'].apply(lambda x : re.findall(regex,x))
```

```python
#ROM
regex = '\d+ GB ROM|\d+ MB ROM|\d+ KB ROM|\d+ TB ROM|NA ROM|NA KB ROM'
df['Specifications'].apply(lambda x : re.findall(regex,x))
df['ROM']=df['Specifications'].apply(lambda x : re.findall(regex,x))
```

```python
df.head()
```

```python
#warranty
regex = '\d Year|\d year'
df['Specifications'].apply(lambda x : re.findall(regex,x))
df['Warranty_in_years']=df['Specifications'].apply(lambda x : re.findall(regex,x))
```

```python
df.head()
```

```python
#RAM
regex = '\d+ GB RAM|\d+ MB RAM|NA RAM|NA MB RAM|NA KB RAM'
df['Specifications'].apply(lambda x : re.findall(regex,x))
df['RAM']=df['Specifications'].apply(lambda x : re.findall(regex,x))
```
